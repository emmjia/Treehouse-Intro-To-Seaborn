# Pokemon Data Visualization

```python
import seaborn as sns
import pandas as pd
```

```python
sns.set_theme()
```

```python
pokemon = pd.read_csv('pokemon_40.csv')
pokemon.head()
```

##Intital Questions

What is the relationship between Attack and HP?

What is the distribution of Attack?

What is the relationship between Attack and Type?

What is the distribution of Attack for each Type?

What is the average (mean) Attack for each Type?

What is the count of Pokemon for each Type?

## What is the Relationship Between Attack and HP?

```python
sns.scatterplot(data=pokemon, x ='HP', y = 'Attack')
```

```python
sns.lineplot(data = pokemon, x = 'HP', y = 'Attack')
```

```python
sns.scatterplot(data = pokemon, x = 'HP', y = 'Attack', hue = 'Type')
```

```python
sns.relplot(data = pokemon, x = 'HP', y = 'Attack', hue = 'Type', col = 'Type')
```

```python
sns.relplot(data = pokemon, x = 'HP', y = 'Attack', hue = 'Type', col = 'Type', col_wrap = 3)
```

Generally, HP and Attack are somewhat positively correlated. Pokemon with more HP, tend to have higher attack.

```python
sns.scatterplot(data = pokemon, x = 'HP', y = 'Defense', hue = 'Type')
```

```python
sns.relplot(data = pokemon, x = 'HP', y = 'Defense', hue = 'Type', col = 'Type', col_wrap = 3)
```

Generally, HP and Defense are somewhat positively correlated. Pokemon with more HP, tend to have higher Defense.

##What is the Distirbution of Attack?

```python
sns.histplot(data = pokemon, x = 'Attack')
```

```python
sns.histplot(data = pokemon, x = 'Attack', bins = 10)
```

```python
sns.kdeplot(data = pokemon, x = 'Attack')
```

```python
sns.histplot(data = pokemon, x = 'Attack', bins = 10, kde = True)
```

```python
sns.displot(data = pokemon, x = 'Attack', bins = 10, col = 'Type', col_wrap = 3 )
```

Most Pokemon have an Attack between 50 and 120.

```python
sns.histplot(data = pokemon, x = 'Defense', bins = 12, kde = True)
```

Pokemon Defense are mostly right skewed. There are two large spikes in defense between 40 and 50 and between 70 and 80.

##What is the relationship between Attack and Type?

```python
sns.stripplot(data = pokemon, x = 'Type', y = 'Attack')
```

```python
sns.catplot(kind = 'strip', data = pokemon, x = 'Type', y = 'Attack', aspect = 2, hue= 'Type')
```

The Dragon types have the highest Attack points, while the Normal types have the largest spread of data.

```python
sns.catplot(kind = 'strip', data = pokemon, x = 'Type', y = 'Defense', aspect = 2, hue= 'Type')
```

After reviewing the Defense data of each Pokemon Type, we can infer that Dragon has the highest Defense points while Normal has the largest spread of data. Very similar inference to our Attack analysis.


## What is the distribution of Attack for the Normal Type?

```python
sns.catplot(kind = 'box', data = pokemon, x = 'Type', y = 'Attack', aspect = 2)
```

```python
sns.catplot(kind = 'violin', data = pokemon, x = 'Type', y = 'Attack', aspect = 2)
```

For the normal type pokemon the minimum is attack 5, the maximum is 110, the median is 75. The first and third quartiles are 55, and 105, respectively.

## What is the distribution of Defense for the Normal Type?

```python
sns.catplot(kind = 'box', data = pokemon, x = 'Type', y = 'Defense', aspect = 2)
```

For normal type pokemon the minimum defense points  is  42, the max 98, the median is 70. The first and third quartiles are 50 and 75 respectively. Also note, there is a outlier at 5.
