## Challenge 2

What is the relationship between:

*   Level and Attack Points?
*   Level and Defense Points?

Bonus: use sns.relplot(kind = 'scatter') to solve one of these questions



```python
sns.lineplot(data = monsters, x ='Level', y = 'Attack_Points')
```

```python
sns.scatterplot(data = monsters, x = 'Level', y = 'Attack_Points', hue = 'Attributes')
```

```python
sns.relplot(data = monsters, x = 'Level', y = 'Attack_Points', hue = 'Attributes', col = 'Attributes', col_wrap = 2)
```

After reviewing a scatterplot and lineplot, we can see that there is a positive correlation between level and attack points.

```python
sns.lineplot(data = monsters, x ='Level', y = 'Defense_Points')
```

```python
sns.scatterplot(data = monsters, x = 'Level', y = 'Defense_Points', hue = 'Attributes')
```

```python
sns.relplot(data = monsters, x = 'Level', y = 'Defense_Points', hue = 'Attributes', col = 'Attributes', col_wrap = 2)
```

In general there is a positive correlation between Level and defense points.

## Challenge 3

What is the distirbution of:

*   Attack Points?
*   Defense Points?

Hint: Use sns.histplot and sns.kdeplot



```python
sns.histplot(data = monsters, x = 'Attack_Points', bins = 10)
```

```python
sns.kdeplot(data = monsters, x = 'Attack_Points')
```

```python
sns.histplot(data = monsters, x = 'Attack_Points', bins = 10, kde = True)
```

After analyzing the histogram and KDE, there seems to be a random distribution of attackpoints. with a large spike around 750 - 1000, some right skewness.

```python
sns.histplot(data = monsters, x = 'Defense_Points', bins = 10)
```

```python
sns.kdeplot(data = monsters, x = 'Defense_Points')
```

```python
sns.histplot(data = monsters, x = 'Defense_Points', bins = 10, kde = True)
```

Defense distribution can be described as random as well with two large spikes at around 1000-1250 and 1750-2000, arguably at around 2750 and 3000 there  is some outliers.

# Challenge 4

For each attribute, what is the relationship between Attribute and:

*   Attack Points?
*   Defense Points?

Hint: use sns.stripplot and sns.swarm plot.

```python
sns.catplot(kind = 'strip', data = monsters, x = 'Attributes', y = 'Attack_Points', aspect = 2, hue= 'Attributes')
```

```python
sns.swarmplot(data = monsters, x = 'Attributes', y = 'Attack_Points')
```

Water and Light have the hightest attack points.  Water and dark have the largest spread in attack points.

```python
sns.catplot(kind = 'strip', data = monsters, x = 'Attributes', y = 'Defense_Points', aspect = 2, hue= 'Attributes')
```

```python
sns.swarmplot(data = monsters, x = 'Attributes', y = 'Defense_Points')
```

Water has the most defense point and the largest spread.
